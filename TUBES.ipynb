KORELASI

- 2018

```python
plt.scatter(Tanah_Longsor_2018, Banjir_2018)
plt.xlabel("Tanah Longsor")
plt.ylabel("Banjir")
plt.title("Analisis Korelasi Tanah Longsor dan Banjir Tahun 2018")
plt.tight_layout()
plt.show()
```

```python
r, p_value = st.pearsonr(Tanah_Longsor_2018, Banjir_2018)

print(r)

if r > 0:
  print("Terdapat korelasi positif antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  if r > 0.5:
    print("Ini adalah korelasi positif yang kuat")
elif r < 0:
  print("Terdapat korelasi Negatif antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  if r < -0.5:
    print("Ini adalah korelasi Negatif yang kuat")
else:
  print("Tidak ada korelasi linier antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  
alpha = 0.05

if p_value < 0.05:
  print(p_value)
  print("Korelasinya signifikan secara statistik.")
else:
  print(p_value)
  print("Korelasinya tidak signifikan secara statistik.")
```

UJI LINEARITAS

- Tanah Longsor dengan Cuaca

- 2018

```python
sns.regplot(x=Cuaca_2018, y=Tanah_Longsor_2018, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('Cek Uji Linear Tanah Longsor dengan Cuaca Tahun 2018')
plt.show()
```

```python
X = Cuaca_2018
X = sm.add_constant(X)
y  = Tanah_Longsor_2018
model = sm.OLS(y, X, missing='drop')
model_result = model.fit()
print(model_result.summary())
```

R-squared:                       0.738

- Asumsi 2
Durbin-Watson:                   1.782

- Asumsi 3
Jarque-Bera (JB):                4.238

```python
plt.scatter(model_result.predict(X), model_result.resid)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Analisis Asumsi Homogenitas Error")
plt.tight_layout()
plt.show()
```

```python
tstat, p_value = stats.levene(Cuaca_2018,Tanah_Longsor_2018)
print('Uji Levene')
if p_value < 0.05:
    print('H0 ditolak, H1 diterima')
else:
    print('H0 diterima')
```

ANOVA ONE WAYS

```python
tahun = ["2018", "2019", "2020", "2021", "2022"]
tanah_longsor = [RataRata_Tanah_Longsor_2018, RataRata_Tanah_Longsor_2019, RataRata_Tanah_Longsor_2020, RataRata_Tanah_Longsor_2021, RataRata_Tanah_Longsor_2022]
banjir = [RataRata_Banjir_2018, RataRata_Banjir_2019, RataRata_Banjir_2020, RataRata_Banjir_2021, RataRata_Banjir_2022]
cuaca = [RataRata_Cuaca_2018, RataRata_Cuaca_2019, RataRata_Cuaca_2020, RataRata_Cuaca_2021, RataRata_Cuaca_2022]

tahun = list(map(int, tahun))
data_dict = {
    "tahun": tahun,
    "tanah_longsor": tanah_longsor,
    "banjir": banjir,
    "cuaca": cuaca
}
df = pd.DataFrame(data_dict)
print(df)
```

```python
df_melted = df.melt(id_vars=["tahun"], value_vars=["tanah_longsor", "banjir", "cuaca"],
                    var_name="kategori", value_name="nilai")

tanah_longsor_data = df_melted[df_melted['kategori'] == 'tanah_longsor']['nilai']
banjir_data = df_melted[df_melted['kategori'] == 'banjir']['nilai']
cuaca_data = df_melted[df_melted['kategori'] == 'cuaca']['nilai']
```

```python
bartlett_stat, bartlett_p_value = bartlett(tanah_longsor_data, banjir_data, cuaca_data)
levene_stat, levene_p_value = levene(tanah_longsor_data, banjir_data, cuaca_data)
anova_stat, anova_p_value = f_oneway(tanah_longsor_data, banjir_data, cuaca_data)

print('Uji Bartlett')
print(f"Statistik Uji: {bartlett_stat}")
print(f"P-value: {bartlett_p_value}")
if bartlett_p_value < 0.05:
    print('H0 ditolak, H1 diterima')
else:
    print('H0 diterima')
    
print('Uji Levene')
print(f"Statistik Uji: {levene_stat}")
print(f"P-value: {levene_p_value}")
if levene_p_value < 0.05:
    print('H0 ditolak, H1 diterima')
else:
    print('H0 diterima')

print('Uji Anova satu arah')
alpha = 0.05
print(f"F-statistik: {anova_stat}")
print(f"P-value: {anova_p_value}")
if anova_p_value < alpha:
    print("H0 ditolak, H1 diterima")
else:
    print("H0 diterima")
```

```python
if anova_p_value < 0.05:
    posthoc_df = sp.posthoc_ttest([tanah_longsor_data, banjir_data, cuaca_data], equal_var=True, p_adjust="bonferroni")

    group_names = ["Tanah Longsor", "Banjir", "Cuaca"]
    posthoc_df.columns = group_names
    posthoc_df.index = group_names

    # Menampilkan hasil post hoc dengan warna latar belakang untuk p-value < 0.05
    print("\nHasil analisis post hoc t-test dengan koreksi Bonferroni:")
else:
    print("\nTidak ada perbedaan signifikan yang terdeteksi oleh uji ANOVA. Analisis post hoc tidak diperlukan.")
```

UJI Shapiro Wilk atau Kolmogorov Smirnov

```python
# Mengelompokkan data berdasarkan kategori (jenis bencana)
tanah_longsor_data = df["tanah_longsor"]
banjir_data = df["banjir"]
cuaca_data = df["cuaca"]

# Uji Normalitas Shapiro-Wilk
shapiro_tanah_longsor = shapiro(tanah_longsor_data)
shapiro_banjir = shapiro(banjir_data)
shapiro_cuaca = shapiro(cuaca_data)

# Uji Normalitas Kolmogorov-Smirnov
ks_tanah_longsor = kstest(tanah_longsor_data, 'norm', args=(tanah_longsor_data.mean(), tanah_longsor_data.std()))
ks_banjir = kstest(banjir_data, 'norm', args=(banjir_data.mean(), banjir_data.std()))
ks_cuaca = kstest(cuaca_data, 'norm', args=(cuaca_data.mean(), cuaca_data.std()))

# Menampilkan hasil uji normalitas
print("\nHasil uji Shapiro-Wilk:")
print(f"Tanah Longsor: Statistik Uji = {shapiro_tanah_longsor[0]}, P-value = {shapiro_tanah_longsor[1]}")
print(f"Banjir: Statistik Uji = {shapiro_banjir[0]}, P-value = {shapiro_banjir[1]}")
print(f"Cuaca: Statistik Uji = {shapiro_cuaca[0]}, P-value = {shapiro_cuaca[1]}")

print("\nHasil uji Kolmogorov-Smirnov:")
print(f"Tanah Longsor: Statistik Uji = {ks_tanah_longsor[0]}, P-value = {ks_tanah_longsor[1]}")
print(f"Banjir: Statistik Uji = {ks_banjir[0]}, P-value = {ks_banjir[1]}")
print(f"Cuaca: Statistik Uji = {ks_cuaca[0]}, P-value = {ks_cuaca[1]}")
```

```python
# Fungsi untuk menentukan hasil uji normalitas
def interpret_normality_test(stat, p_value, alpha=0.05):
    if p_value < alpha:
        print(f"Statistik Uji = {stat}, P-value = {p_value} (H0 diterima: data berdistribusi normal)")
    else:
        print(f"Statistik Uji = {stat}, P-value = {p_value} (H0 ditolak: data tidak berdistribusi normal)")
```

```python
interpret_normality_test(shapiro_tanah_longsor[0], shapiro_tanah_longsor[1])
interpret_normality_test(shapiro_banjir[0], shapiro_banjir[1])
interpret_normality_test(shapiro_cuaca[0], shapiro_cuaca[1])

interpret_normality_test(ks_tanah_longsor[0], ks_tanah_longsor[1])
interpret_normality_test(ks_banjir[0], ks_banjir[1])
interpret_normality_test(ks_cuaca[0], ks_cuaca[1])
```
