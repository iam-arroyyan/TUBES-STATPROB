KORELASI

- 2018

```python
plt.scatter(Tanah_Longsor_2018, Banjir_2018)
plt.xlabel("Tanah Longsor")
plt.ylabel("Banjir")
plt.title("Analisis Korelasi Tanah Longsor dan Banjir Tahun 2018")
plt.tight_layout()
plt.show()
```

```python
r, p_value = st.pearsonr(Tanah_Longsor_2018, Banjir_2018)

print(r)

if r > 0:
  print("Terdapat korelasi positif antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  if r > 0.5:
    print("Ini adalah korelasi positif yang kuat")
elif r < 0:
  print("Terdapat korelasi Negatif antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  if r < -0.5:
    print("Ini adalah korelasi Negatif yang kuat")
else:
  print("Tidak ada korelasi linier antara Banjir dan Tanah Longsor Pada Tahun 2018.")
  
alpha = 0.05

if p_value < 0.05:
  print(p_value)
  print("Korelasinya signifikan secara statistik.")
else:
  print(p_value)
  print("Korelasinya tidak signifikan secara statistik.")
```

UJI LINEARITAS

- Tanah Longsor dengan Cuaca

- 2018

```python
sns.regplot(x=Tanah_Longsor_2018, y=Cuaca_2018, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('Cek Uji Linear Tanah Longsor dengan Cuaca Tahun 2018')
plt.show()
```

```python
X = Tanah_Longsor_2018
X = sm.add_constant(X)
y  = Cuaca_2018
model = sm.OLS(y, X, missing='drop')
model_result = model.fit()
print(model_result.summary())
```

R-squared:                       0.738

- Asumsi 2
Durbin-Watson:                   1.782

- Asumsi 3
Jarque-Bera (JB):                4.238

```python
plt.scatter(model_result.predict(X), model_result.resid)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Analisis Asumsi Homogenitas Error")
plt.tight_layout()
plt.show()
```

```python
tstat, p_value = stats.levene(Tanah_Longsor_2018,Cuaca_2018)
print('Uji Levene')
if p_value < 0.05:
    print('H0 ditolak, H1 diterima')
else:
    print('H0 diterima')
```

ANOVA

- Tanah Longsor dengan Cuaca

```python
# DATA HARUS BERBENTUK STACK
Bencana = ['TanahLongsor', 'TanahLongsor', 'TanahLongsor', 'TanahLongsor', 'TanahLongsor',
            'Cuaca', 'Cuaca', 'Cuaca', 'Cuaca', 'Cuaca']
Tahun = [ '2018', '2019', '2020', '2021', '2022',
            '2018', '2019', '2020', '2021', '2022', ]
Jumlah_Bencana = [Tanah_Longsor_2018.sum(), Tanah_Longsor_2019.sum(), Tanah_Longsor_2020.sum(), Tanah_Longsor_2021.sum(), Tanah_Longsor_2022.sum(),
                    Cuaca_2018.sum(), Cuaca_2019.sum(), Cuaca_2020.sum(), Cuaca_2021.sum(), Cuaca_2022.sum()]
#jadikan pandas
data = pd.DataFrame()
data["Bencana"]=Bencana
data["Tahun"]=Tahun
data["Jumlah"]=Jumlah_Bencana
```

```python
tstat, p_value = stats.levene(data['Jumlah'][data['Bencana'] == 'TanahLongsor'],
                              data['Jumlah'][data['Bencana'] == 'Cuaca'])
print('Uji Levene')
if p_value < 0.05:
    print('H0 ditolak, H1 diterima')
else:
    print('H0 diterima')
```

```python
model = ols("Jumlah ~ Bencana + Tahun + Bencana:Tahun",data = data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
print(anova_table)
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
fig = interaction_plot(
    x= data["Tahun"],
    trace= data["Bencana"],
    response= data["Jumlah"],
    colors=["red", "blue"],
    markers=["D", "^"],
    ms=10,
    ax=ax,
)
```
